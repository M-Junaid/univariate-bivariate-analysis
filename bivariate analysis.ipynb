## 4. Distplot (Numerical - Categorical)

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'],hist=False)
sns.distplot(titanic[titanic['Survived']==1]['Age'],hist=False)
sns.set_theme()

# sns.distplot(df["Age"],hist=False)
```

```python
print(flights.dtypes)
```

```python
numeric_columns = flights.select_dtypes(include=['int64', 'float64']).columns
result = flights.groupby("year")[numeric_columns].sum().reset_index()

```

```python
result
```

```python
sns.lineplot(x='year', y='passengers', data=result)
```
